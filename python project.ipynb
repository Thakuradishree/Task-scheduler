```python
import pandas as pd
import datetime
import smtplib,ssl
from email.message import EmailMessage
import os
import schedule
import time
from time import sleep

def birthday():
    def sendEmail(to, sub, msg):
        print(f"email to {to} \nsend with subject: {sub}\n message: {msg}")
        email = EmailMessage()
        email['from'] = 'Jiya Rathi'
        email['to'] = f"{to}"
        email['subject'] = f'{sub}'

        email.set_content(f'{msg}')

        with smtplib.SMTP(host='smtp.gmail.com', port=587) as smtp:
            smtp.ehlo()
            smtp.starttls()
            smtp.login('ajjer.2002@gmail.com','humhaijiya14')
            smtp.send_message(email)
            print("Email send")
        pass
    if __name__ == "__main__":
        df = pd.read_excel("data.xlsx")
        print(df)
        today = datetime.datetime.now().strftime("%d-%m")
        #print(type(today))
        update = []
        yearnow =  datetime.datetime.now().strftime("%Y")
        #print(yearnow)
        for index, item in df.iterrows():
            #print(index,item['birthday'])
            bday = item['Birthday'].strftime("%d-%m")
            #print(type(bday))
            if(bday == today) and yearnow not in str(item["Year"]): 
                sendEmail(item['Email'] ,"Happy BIrthday "+item["Name"], item['message'])
                update.append(index)
        for i in update:
            yr = df.loc[i, 'Year']
            #print(yr)
            df.loc[i,'Year'] = f"{yr}, {yearnow}"
            #print((df.loc[i, 'Year'])
        #print(df)
        df.to_excel("data.xlsx", index=False)

def tasking():
    email=input("Enter a valid email: ")
    n= int(input("Enter the number of time stamps you want to schedule tasks for : "))
    print("\n")
    print("-------------------------------")
    events= {}
    message=""

    for x in range(0,n):
        print("TASK" ,x+1)
        key=input("Enter the time for event: " )
        value = input("Enter the event to be scheduled: ") 
        print("Time : ",key)
        print("Task : ",value)
        events[key] = value
        
        print("-------------------------------")

    events_list=[]
    et=0

    for i in events:
        k=(i,events[i])
        events_list.append(k)

    def mailing():
    
        # creates SMTP session
        s = smtplib.SMTP('smtp.gmail.com', 587)
    
        # start TLS for security
        s.starttls()
    
        # Authentication
        s.login("jiyarathi546@gmail.com", "humhaijiya14")
        
        #sending message
        message=events_list[et][1]

        # sending the mail
        s.sendmail("jiyarathi546@gmail.com", email, message)
    
        # terminating the session
        s.quit()
        
    def mail_timer():
        mailing()
        global et
        et+=1

    for x in events:
        schedule.every().day.at(x).do(mail_timer)

    while True:
        schedule.run_pending()
        sleep(1)

username = input("Give me your username: ")
print("Your name is " + username)

flag=1

def switch():
    global flag

    print("\nWhat would you like to do today?")

    print("1. Set reminder for birthdate of a friend: ")
    print("2. Set reminder for today's task") 
    print("3. Exit")
        
    option = int(input("Enter your choice : "))

    if option == 1:
        birthday()   
        
    elif option == 2:
       tasking()
        
    elif option == 3:
        flag=-1
        print("Execution terminated")
    else:
        print("Choose the correct option")

        
while flag>0:
    switch()
```
